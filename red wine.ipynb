Defining X and Y

```python
# Define X and y
X = dataset.iloc[:,:-1]  # All rows, first 11 columns
# Check dimensions of X 
X.shape
```

```python
# label binarization
y = (dataset['quality'] >= 6).astype(int)
print(y)
```

```python
#Visulisation of the y
import seaborn as sns
import matplotlib.pyplot as plt
# Create the count plot
ax = sns.countplot(x=y)
plt.title('Distribution of Wine Quality Ratings')
plt.xlabel('High Quality (1 = Yes, 0 = No)')
plt.ylabel('Count')
# Annotate the bars with the count
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points')

plt.show()
```

```python
# Split data into train and test sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, 
test_size=0.6, random_state=42, stratify=y)
```

```python
print("X_train shape:", X_train.shape)
print("y_train shape:", y_train.shape)
print("X_test shape:", X_test.shape)
print("y_test shape:", y_test.shape)
```

Scaling

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train_s=sc.fit_transform(X_train)
X_test_s=sc.transform(X_test)
```

Feature selection

```python
from sklearn.feature_selection import VarianceThreshold 
variance_selector = VarianceThreshold(threshold=0)
X_train_fs = variance_selector.fit_transform(X_train_s) 
X_test_fs = variance_selector.transform(X_test_s)
print(f"{X_train.shape[1]-X_train_fs.shape[1]} features have been removed, {X_train_fs.shape[1]} features remain")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Getting the boolean mask of the selected features from VarianceThreshold
selected_mask = variance_selector.get_support()

# Creating a DataFrame for visualization
features_df = pd.DataFrame({
    'Feature': X_train.columns,
    'Retained': selected_mask
})

# Convert the boolean mask to integer (1 for retained, 0 for dropped)
features_df['Retained'] = features_df['Retained'].astype(int)

# Plotting the retained and dropped features
plt.figure(figsize=(12, 6))
sns.barplot(x='Feature', y='Retained', data=features_df)
plt.xticks(rotation=90)
plt.title('Retained and Dropped Features after Variance Thresholding')
plt.ylabel('Retained (1 = Yes, 0 = No)')
plt.xlabel('Feature')
plt.show()
```

```python
# using the SMOTE METHOD
from imblearn.over_sampling import SMOTE
import seaborn as sns
from imblearn.over_sampling import SMOTE

# Set k_neighbors to a smaller number, based on your smallest class size
smote = SMOTE(random_state=42, k_neighbors=1)  
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)

# Now to visualize the new class distribution
plt.figure(figsize=(10, 6))
sns.countplot(x=y_train_smote)
plt.title('Class Distribution after SMOTE')
plt.xlabel('Quality')
plt.ylabel('Count')
plt.xticks(rotation=45)  # Rotate x-labels if they overlap
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier

rf_classifier = RandomForestClassifier(random_state=42)
```

```python
rf_classifier.fit(X_train_fs, y_train)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
# Make predictions on the test data
y_pred = rf_classifier.predict(X_test_fs)
print(f"Accuracy Score: {accuracy_score(y_test,y_pred)*100:.2f}%") 
cm = confusion_matrix(y_test,y_pred)
ax = sns.heatmap(cm, cmap='flare',annot=True, fmt='d')
plt.xlabel("Predicted Class",fontsize=12) 
plt.ylabel("True Class",fontsize=12) 
plt.title("Confusion Matrix",fontsize=12)
plt.show()
```

Hyperparameter in KNN

```python

from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn import metrics

# Assuming X_train_fs, X_test_fs, y_train, y_test are already defined and preprocessed

# Initialize the classifier with default parameters
classifier = KNeighborsClassifier()

# Define the hyperparameter grid to search
param_grid = {
    'n_neighbors': range(1, 31, 2),  # testing odd values to avoid ties
    'metric': ['euclidean', 'manhattan', 'minkowski'],
    'p': [1, 2]  # 1 is manhattan distance, 2 is euclidean distance when metric is minkowski
}

# Set up the grid search with cross-validation
grid_search = GridSearchCV(
    estimator=classifier,
    param_grid=param_grid,
    cv=5,
    n_jobs=-1,
    verbose=1,
    scoring='accuracy'
)

# Perform the grid search to find the best hyperparameters
grid_search.fit(X_train_fs, y_train)

# Print out the best parameters found by the grid search
print("Best parameters found: ", grid_search.best_params_)

# Use the best estimator to make predictions on the test set
y_pred = grid_search.predict(X_test_fs)

# Evaluate and print the accuracy
acc = metrics.accuracy_score(y_test, y_pred)
print('Accuracy: %.2f\n\n' % acc)

# Print the confusion matrix
cm = metrics.confusion_matrix(y_test, y_pred)
print('Confusion Matrix:')
print(cm, '\n\n')

# Print the classification report
result = metrics.classification_report(y_test, y_pred)
print('Classification Report:\n')
print(result)
```
